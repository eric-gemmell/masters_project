```python
import Bio as bio
from Bio import SeqIO
from Bio import Phylo
from Bio import AlignIO
from Bio.Align import MultipleSeqAlignment
from Bio.Phylo.TreeConstruction import DistanceCalculator
from Bio.Phylo.TreeConstruction import DistanceTreeConstructor
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
import matplotlib
import matplotlib.pyplot as plt

    #Check out this resource https://biopython.org/wiki/PAML
```

```python
raw_fasta = SeqIO.parse("raw_sequences/Azoline/Bottromycin/seqdump.txt", "fasta")

sequences = [seq for seq in raw_fasta]
aa_sequences = [str(seq.seq) for seq in sequences]

max_len = max([len(aa_seq) for aa_seq in aa_sequences])


padded_sequences = [aa_seq.ljust(max_len, '-') for aa_seq in aa_sequences]
for i in range(0,len(sequences)):
    sequences[i].seq = Seq(padded_sequences[i])
    
SeqIO.write(sequences, "padded.fasta", "fasta")
```

```python
align = AlignIO.read("padded.fasta", "fasta")

calculator = DistanceCalculator('identity')
distance_matrix = calculator.get_distance(align)
print(distance_matrix)
constructor = DistanceTreeConstructor(calculator)

tree = constructor.build_tree(align)
tree.rooted = True
fig = Phylo.draw(tree)
```
